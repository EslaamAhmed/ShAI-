## Data cleaning 

```python
print("num of duplicates: "+ str(df.duplicated().sum()))
print(df.isnull().sum())
# all values in NOTES & Status are null so will be deleted later 
```

```python
df[df['TotalPay'] == 0].head(10)
df[df['TotalPay'] == 0].tail(10)
# no important data related to salary in this rows so will be deleted 
```

```python
df=df[df['TotalPay']>0] # to drop all rows where salary <= 0 
df.drop(['Notes','Status'], axis=1, inplace=True)
```

```python
df[df['Benefits'].isnull()]
```

im gonna substitute Null value up on this Rules<br>
BasePay + OvertimePay + otherPay = TotalPay  <br>
TotalPay + Benefits = TotalPayBenefits 

BasePay = TotalPay - OtherPay - OverTime<br>
Benefits = TotalPayBenfits - TotalPay

```python
df[df['BasePay'].isnull()].head()
```

```python
df['BasePay'].fillna(df['TotalPay'] - df['OtherPay'] - df['OvertimePay'], inplace=True) #replace  NaN with the difference between Total Pay and Other Pay + Overtime Pay
df['Benefits'].fillna(df['TotalPayBenefits'] - df['TotalPay'], inplace=True) #replace  NaN with Total Pay Benefits minus Total Pay
```

## Grouped analysis

```python
salary_year=df.groupby('Year')['TotalPay'].agg(['mean','min','max'])

fig, axs = plt.subplots(1,3,figsize=(15, 5))
fig.suptitle('comparing salaries in different years')
axs[0].bar(salary_year.index, salary_year['mean'].values)
axs[0].set_xlabel('Year')
axs[0].set_ylabel('Salary')
axs[0].set_title('Avg. Salary by Year')
axs[0].bar_label(axs[0].containers[0])

axs[1].bar(salary_year.index, salary_year['max'].values)
axs[1].set_xlabel('Year')
axs[1].set_title('Max. Salary by Year')
axs[1].bar_label(axs[1].containers[0])

axs[2].bar(salary_year.index, salary_year['min'].values)
axs[2].set_xlabel('Year')
axs[2].set_title('Min. Salary by Year')
axs[2].bar_label(axs[2].containers[0])


plt.show()
```

```python
job_salary=df.groupby('Year').max()[['TotalPay','JobTitle']]
plt.bar(job_salary.index,job_salary['TotalPay'])
plt.xticks(job_salary.index)
plt.xlabel('Year')
plt.ylabel('Total Pay')
plt.title('Max Total Pay by Year and Job Title')
for year, row in job_salary.iterrows():
    plt.text(year, row['TotalPay'], row['JobTitle'], ha='center')
plt.show()
```

## Simple Correlation Analysis

```python
numeric_df = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix=numeric_df.drop(columns=['Id']).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
